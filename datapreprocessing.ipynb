## Data Combination

### food nutrient data

#### Merge food.csv and food_nutrient.csv

```python
# Use fdc_id as common column
food_nutrient_processed = pd.merge(food_nutrient, food, on="fdc_id")
# Display head() of combined data
food_nutrient_processed.head()
```

#### Organize Columns orders and Exclude Unused Columns

```python
# data_points, min, max, median, footnote, min_year_acquired, publication_date were excluded

columnsTitles = ['id', 'fdc_id', 'nutrient_id', 'description', 'amount', 'derivation_id', 'data_type', 'food_category_id']

food_nutrient_processed = food_nutrient_processed.reindex(columns=columnsTitles)

food_nutrient_processed.head()
```

#### Add nutrient names using nutrient.csv

```python
# Replace id column with common name nutrient_id
nutrient = nutrient.rename(columns = {'id': 'nutrient_id'})
nutrient.head()
```

```python
# Use nutrient_id as common column
food_nutrient_processed = pd.merge(food_nutrient_processed, nutrient, on="nutrient_id")
# Display head() of combined data
food_nutrient_processed.head()
```

#### Export to csv File

```python
food_nutrient_processed.to_csv('food_nutrient_processed.csv')
```

### Food Protein, Fat and Carbohydrate Data

#### Merge food_calorie_conversion_factor.csv and food_nutrient_conversion_factor.csv

```python
nutrient_factor = nutrient_factor.rename(columns = {'id':'food_nutrient_conversion_factor_id'})
food_factor = pd.merge(calorie_factor, nutrient_factor, on="food_nutrient_conversion_factor_id")
food_factor.head()
```

#### Merge food_factor and food.csv

```python
food_protein = pd.merge(food, food_factor, on="fdc_id")
food_protein.head()
```

```python
# publication_date, food_nutrient_conversion_factor_id

columnsTitles = ['fdc_id', 'data_type', 'description', 'food_category_id', 'protein_value', 'fat_value', 'carbohydrate_value']
food_protein = food_protein.reindex(columns=columnsTitles)
food_protein.head()
```

#### Export to csv File

```python
food_protein.to_csv('food_protein.csv')
```
